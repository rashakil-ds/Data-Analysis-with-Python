# Sinx and Cosx

```python
import numpy as np

x = np.linspace(0, 4*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(10,6))
plt.plot(x, y1, label='Sin(x)', linestyle='-', color='blue', linewidth=2, marker='o', markersize=8)
plt.plot(x, y2, label='Cos(x)', linestyle='--', color='green', linewidth=2, marker='s', markersize=8)

plt.axhline(y=0, color='green', linestyle='--', linewidth=1)
plt.legend(loc='upper right', shadow=True)

plt.grid(True, linestyle='--', alpha=0.2)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')

plt.savefig('custom_line_plot.png')

plt.show()
```
