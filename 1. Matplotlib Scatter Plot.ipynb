Matplotlib is a versatile library with a wide range of features for creating static, animated, and interactive visualizations. Here's a list of some key features:

1. **Line Plots:** Create simple line plots to visualize the relationship between two variables.

2. **Scatter Plots:** Display individual data points on a graph to observe patterns or relationships.

3. **Bar Plots:** Represent categorical data using bars, suitable for comparing different categories.

4. **Histograms:** Visualize the distribution of a dataset by dividing it into bins and displaying the frequency of values in each bin.

5. **Pie Charts:** Represent data as slices of a circular pie, useful for displaying the proportion of different categories in a whole.

6. **Error Bars:** Add error bars to indicate uncertainty or variability in data points.

7. **Annotations:** Annotate plots with text, arrows, or other shapes to provide additional information.

8. **Customization:** Customize almost every aspect of a plot, including colors, line styles, markers, fonts, axis labels, titles, and more.

9. **Subplots:** Create multiple plots within the same figure to compare different datasets or aspects of data.

10. **3D Plotting:** Generate 3D plots to visualize data in three dimensions.

11. **Interactive Features:** Matplotlib can be used with interactive backends, allowing users to interact with and explore plots.

12. **Animations:** Create animated visualizations to represent changes in data over time.

13. **Seaborn Integration:** Matplotlib can be combined with Seaborn, a statistical data visualization library based on Matplotlib, to create aesthetically pleasing and informative visualizations.

14. **Latex Support:** Matplotlib supports LaTeX for rendering mathematical expressions in labels and titles.

15. **Saving Plots:** Save plots in various formats, including PNG, PDF, SVG, and more.

16. **Integration with NumPy:** Matplotlib seamlessly integrates with NumPy, allowing users to use NumPy arrays as input data.

17. **Backend Support:** Matplotlib supports different backends, including interactive backends for GUI applications and non-interactive backends for saving plots without displaying them.

18. **Themes and Styles:** Apply predefined styles or create custom styles to control the overall appearance of plots.

19. **Color Maps:** Choose from a variety of color maps to enhance the visual appeal of plots.

20. **Polar Plots:** Create plots in polar coordinates for circular or angular data.

These features make Matplotlib a powerful tool for data visualization in Python, suitable for a wide range of applications and domains. The library's extensive documentation and active community support further contribute to its popularity.

```python
x = [1,2,3,4,19,30]
y = [3,6,9,10,15,20]

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
```

# Sinusoidal Curve

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)

plt.axhline(y=0, color='green', linestyle='--', linewidth=1)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sinusoidal Curve with Horizontal Line at y=0')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)

plt.axhline(y=0, color='green', linestyle='--', linewidth=1)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sinusoidal Curve with Horizontal Line at y=0')
```
