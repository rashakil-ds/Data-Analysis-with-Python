# Boxplot with Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.array([1,2,3,3,5,7,7,8,9,10,20])
q1 = np.percentile(data, 25) # 25% percentile/ Q1
q3 = np.percentile(data, 75) # 75% percentile/ Q3
iqr = q3 - q1 #inter quartile range

upper_whisker = q3 + (1.5 * iqr) # Q3 + (1.5*IQR)
lower_whisker = q1 - (1.5 * iqr) #Q1 – (1.5*IQR)


outliers = (data < lower_whisker) | (data > upper_whisker)

# Create the boxplot
fig, ax = plt.subplots()
ax.boxplot(data)

# Show the plot
ax.set_title('Box Plot Example')
ax.set_xlabel('Data')
ax.set_ylabel('Values')
plt.show()
```

# Boxplot with Seaborn

```python
import seaborn as sns
# Sample dataset
data_points = [1,2,3,3,5,7,7,8,9,10,20]

# Create a box plot
sns.boxplot(data = data_points, width=0.1)

# Add labels and title
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Display the plot
plt.show()
```
