```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 5, 2  # mean and std dhore nisi, data generate korbo
mu_std, sigma_std = 0, 1  # mean and std eta constant as its standard
```

# Generate points on the x axis

```python
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
x_std = np.linspace(mu_std - 4*sigma_std, mu_std + 4*sigma_std, 1000)
```

# Calculate the Normal Distribution and Standard Normal Distribution

```python
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2) #PDF
y_std = (1 / (sigma_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_std - mu_std) / sigma_std) ** 2)
```

# Plotting Normal Distribution

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Normal Distribution\n$\mu=5$, $\sigma=2$')
plt.fill_between(x, y, alpha=0.2)
plt.title('Normal Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

# Standard Normal Distribution

```python
plt.figure(figsize=(10, 5))
plt.plot(x_std, y_std, label='Standard Normal Distribution\n$\mu=0$, $\sigma=1$', color='orange')
plt.fill_between(x_std, y_std, alpha=0.2, color='orange')
plt.title('Standard Normal Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

# Calculate PDF

```python
from scipy.stats import norm
```

```python
points_normal = [2, 5, 7, 12]  # Points for the normal distribution
points_standard = [-2, 0, 2, 3]  # Points for the standard normal distribution

pdf_values_normal = [norm.pdf(x, mu, sigma) for x in points_normal]
pdf_values_standard = [norm.pdf(z, mu_std, sigma_std) for z in points_standard]

pdf_results_normal = dict(zip(points_normal, pdf_values_normal))
pdf_results_standard = dict(zip(points_standard, pdf_values_standard))
```

```python
pdf_results_normal
```

```python
pdf_results_standard
```

# Plot for the normal distribution with highlighted points

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution\n$\mu=5$, $\sigma=2$')
plt.fill_between(x, y, alpha=0.2)
plt.scatter(points_normal, [norm.pdf(point, mu, sigma) for point in points_normal], color='red')

for point in points_normal:
    plt.annotate(f"({point}, {norm.pdf(point, mu, sigma):.2f})", (point, norm.pdf(point, mu, sigma)), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Normal Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

# Plot for the standard normal distribution with highlighted points

```python
plt.figure(figsize=(10, 6))
plt.plot(x_std, y_std, label='Standard Normal Distribution\n$\mu=0$, $\sigma=1$', color='orange')
plt.fill_between(x_std, y_std, alpha=0.2, color='orange')
plt.scatter(points_standard, [norm.pdf(point, mu_std, sigma_std) for point in points_standard], color='red')

for point in points_standard:
    plt.annotate(f"({point}, {norm.pdf(point, mu_std, sigma_std):.3f})", (point, norm.pdf(point, mu_std, sigma_std)), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Standard Normal Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```
