**Bar Plot:** Bar plots are used for visualizing the distribution of *categorical data* or the relationship between two categorical variables. The x-axis of a bar plot represents the categories, and the y-axis represents the value associated with each category. Each bar represents a different category, and the height of the bar corresponds to the value.
Bar plots are used when you have distinct categories and want to compare their values.

```python
import matplotlib.pyplot as plt

categories = ['AI', 'Data Science', 'ML', 'DeepML']
values = [20, 35, 30, 55]

plt.bar(categories, values, color='blue')
```

```python
plt.barh(categories, values, color='blue')
```

```python
import matplotlib.pyplot as plt

categories = ['AI', 'Data Science', 'ML', 'DeepML']
values = [20, 35, 30, 55]

plt.bar(categories, values, color='blue')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
```

```python
categories = ['AI', 'Data Science', 'ML', 'DeepML']
values = [20, 35, 30, 55]

#plt.bar(categories, values, color='blue')

fig, ax = plt.subplots()
bars = ax.bar(categories, values, color='orange')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')

plt.xlabel('Categories')
plt.ylabel('Quantity')
plt.title('Bar Plot with Quantity and Values')

plt.show()
```
